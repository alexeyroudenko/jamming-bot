# Noun Phrases, Verbs

````python
import spacy
nlp = spacy.load("en_core_web_lg")
doc = nlp(text)

# длинные, содержательные именные группы текста
noun_phrases =  [chunk.text for chunk in doc.noun_chunks]
noun_phrases_out = []
for i in noun_phrases:
    if len(i) > 5:
        noun_phrases_out.append(i)
noun_phrases =  noun_phrases_out

# чистый список действий/процессов
verbs = [token.lemma_ for token in doc if token.pos_ == "VERB"]

# Find named entities, phrases and concepts
entities = []
for entity in doc.ents:
    entities.append({entity.label_:entity.text})

print("Noun phrases")
print(noun_phrases)
print("Verbs")
print(verbs)
print("Entities")
print(entities)
````

````python
import spacy

nlp = spacy.load("en_core_web_lg")
doc = nlp(text)

# Ищем токен "Jamming bot"
query_token = None
for token in doc:
    if token.text.lower() == "jammingbot":
        query_token = token

if query_token:
    print(f"Найден токен '{query_token.text} {query_token.nbor(1).text}'")

    # Слова, связанные с "Jamming bot"
    print("\nСлова, связанные с 'Jamming bot', и типы зависимостей:")
    for child in query_token.children:
        print(f"{child.text} ({child.dep_})")

    # Родитель "Jamming bot"
    print(f"\nРодитель 'Jammingbot': {query_token.head.text} ({query_token.head.dep_})")

    # Полное предложение
    print("\nПредложение:")
    print(query_token.sent)
else:
    print("'Jamming bot' не найдено в тексте.")
````

# Keywords

````python
import spacy
from collections import Counter

nlp = spacy.load("en_core_web_lg")
doc = nlp(text)
keywords = [token.text.lower() for token in doc if token.pos_ in ["NOUN", "ADJ"] and not token.is_stop]
keyword_counts = Counter(keywords)

print("Top-5 Keywords:")
print(keyword_counts.most_common(5))
````

# Similarity

````python
import spacy
nlp = spacy.load("en_core_web_lg")

doc = nlp(text)
query = nlp("bot")
similarities = {}
for token in doc:
    if token.has_vector and query.has_vector:
        similarity = query.similarity(token)
        if similarity > 0:  # Фильтрация значений, близких к нулю
            similarities[token.text] = similarity

sorted_similarities = sorted(similarities.items(), key=lambda x: x[1], reverse=True)

print(f"Similarity w '{query.text}':")
for word, similarity in sorted_similarities[0:30]:
    print(f"{word}: {similarity:.2f}")
````

````python
import spacy
nlp = spacy.load("en_core_web_lg")
def analyze(text):
    doc = nlp(text)
    def analyze_data(doc):
        noun_phrases = [chunk.text for chunk in doc.noun_chunks]
        verbs = [token.lemma_ for token in doc if token.pos_ == "VERB"]
        entities = [(ent.text, ent.label_) for ent in doc.ents]
        keywords = [token.text.lower() for token in doc if token.pos_ in ["NOUN", "ADJ"] and not token.is_stop]
        subj = [token.text for token in doc if token.dep_ == "nsubj"]
        obj = [token.text for token in doc if token.dep_ == "dobj"]
        dependency = [(token.text, token.dep_) for token in doc]
        return {
            "noun_phrases": noun_phrases,
            "verbs": verbs,
            "entities": entities,
            "keywords": keywords,
            "subjects": subj,
            "objects": obj,
            "dependency": dependency
        }

````
